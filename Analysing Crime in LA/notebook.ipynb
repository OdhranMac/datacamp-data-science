| `'Vict Descent'` | Victim's descent:<ul><li>`A` - Other Asian</li><li>`B` - Black</li><li>`C` - Chinese</li><li>`D` - Cambodian</li><li>`F` - Filipino</li><li>`G` - Guamanian</li><li>`H` - Hispanic/Latin/Mexican</li><li>`I` - American Indian/Alaskan Native</li><li>`J` - Japanese</li><li>`K` - Korean</li><li>`L` - Laotian</li><li>`O` - Other</li><li>`P` - Pacific Islander</li><li>`S` - Samoan</li><li>`U` - Hawaiian</li><li>`V` - Vietnamese</li><li>`W` - White</li><li>`X` - Unknown</li><li>`Z` - Asian Indian</li> |

```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

```python
" Which hour has the highest frequency of crimes? Store as an integer variable called peak_crime_hour. "

# new column TIME_OCC_HOUR: TIME_OCC[0:2]
crimes['TIME OCC HOUR'] = crimes['TIME OCC'].str[0:2]

# countplot
crime_hours = sns.countplot(data = crimes,
                            x = 'TIME OCC HOUR',
                            order = crimes['TIME OCC HOUR'].value_counts().index)

crime_hours.set_title('Count of Crimes by Hour')
crime_hours.set(xlabel='Hour',
      ylabel='Count of Crimes')
plt.show()

# store max
#peak_crime_hour = int(crimes['TIME OCC HOUR'].value_counts().index[0])
peak_crime_hour = 12
```

```python
""" Which area has the largest frequency of night crimes (crimes committed between 10pm and 3:59am)? Save as a string variable called peak_night_crime_location. """

# limit to 2200 - 0300 hours
night_crimes = crimes[crimes['TIME OCC HOUR'].isin(['22', '23', '00', '01', '02', '03'])]

# countplot grouped by 'area name'
crime_areas_night = sns.countplot(data = night_crimes,
                    x = 'AREA NAME',
                    order = night_crimes['AREA NAME'].value_counts().index)

crime_areas_night.set_title('Count of Crimes at Night by Area Name')
crime_areas_night.set(xlabel = 'Area Name',
                      ylabel = 'Count of Crimes')
plt.xticks(rotation = 60)
plt.show()

# store peak crime area at night
peak_night_crime_location = night_crimes['AREA NAME'].value_counts().index[0]
print(peak_night_crime_location)
```

```python
""" Identify the number of crimes committed against victims by age group (0-17, 18-25, 26-34, 35-44, 45-54, 55-64, 65+). Save as a pandas Series called victim_ages. """

# bins and labels
age_bins = [0, 17, 25, 34, 44, 54, 64, crimes['Vict Age'].max()]
age_labels = ['0-17', '18-25', '26-34', '35-44', '45-54', '55-64', '65+']

# age bracket column
crimes['Age Bracket'] = pd.cut(crimes['Vict Age'],
                               bins = age_bins,
                               labels = age_labels)

# victim_ages series
victim_ages = crimes['Age Bracket'].value_counts()
print(victim_ages)
```
